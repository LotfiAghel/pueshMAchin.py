```python
# python ~/a.py -p /dev/ttyACM0 115200 -l 0.0.0.0 --access-list='192.168.44.1'
import socket
import time
import json

def send_command(cmd, noprint=False):
    cmd = cmd + '\n'
    s.sendall(cmd.encode())
    z = ''
    while True:
        try:
            data = s.recv(1024)
        except:
            break
        if not data:
            break
        z = data.decode()[:-1]
        if not noprint:
            print(z)
    return json.loads(z)


HOST = '192.168.44.100'  # The server's hostname or IP address
PORT = 9100        # The port used by the server
s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.connect((HOST, PORT))
s.settimeout(.5)
```
